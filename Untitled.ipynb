```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report
import re
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
fake = pd.read_csv("data/Fake.csv")
fake
```

```python
fake["label"] = 1
fake
```

```python
real = pd.read_csv("data/True.csv")
```

```python
real
```

```python
real["label"] = 0
real
```

```python
data = pd.concat([real, fake ],axis=0)
```

```python
data
```

```python
data = data.sample(frac=1)
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
data = data.drop(["title","subject","date"],axis=1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
def word_drop(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W" ," ", text)
    text = re.sub('https?://\S+|www.\.\S+','',text)
    text = re.sub('<.*?>+' , '',text)
    text = re.sub('\n','',text)
    text = re.sub('\w*\d\w','',text)
    return text
```

```python
data['text'] = data['text'].apply(word_drop)
```

```python
sample_data = data.head(10).to_csv("sample.csv")
sample_data
```

```python
x = data["text"]
y = data["label"]
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size=.25)
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

##### Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
```

```python
LRmodel = LogisticRegression()
```

```python
LRmodel.fit(xv_train ,y_train  )
LRmodel.score(xv_test ,y_test  )
```

```python
Pred_LR = LRmodel.predict(xv_test)
print(classification_report(y_test,Pred_LR))
```

###### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DTmodel = DecisionTreeClassifier()
DTmodel.fit(xv_train ,y_train)
DTmodel.score(xv_test ,y_test  )
```

```python
Pred_DT = DTmodel.predict(xv_test)
print(classification_report(y_test,Pred_DT))
```

###### Gradient Boost 

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GBmodel = GradientBoostingClassifier(random_state=0)
GBmodel.fit(xv_train ,y_train)
GBmodel.score(xv_test ,y_test  )
```

```python
Pred_GB = GBmodel.predict(xv_test)
print(classification_report(y_test,Pred_GB))
```

###### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFmodel = RandomForestClassifier(random_state=0)
RFmodel.fit(xv_train ,y_train)
RFmodel.score(xv_test ,y_test  )
```

```python
Pred_RF = RFmodel.predict(xv_test)
print(classification_report(y_test,Pred_RF))
```
